## I will create a Python function that returns the energy from input masses for the mass energy equivalence. Here, I aslo make plots for a line, quadratic, cubic, and an exponential function.  

### Import packages

```python
import numpy as np

import matplotlib.pyplot as plt
```

## Activity 1: E = mc^2

### Create a python function for the equation E =mc^2  that returns the energy

```python
c = 3*10**8  # [m/s]
def Equiv(n):
    m = np.arange(n)+1
    E = m*c**2
    return E
```

```python
Energy = Equiv(15)
mass = np.arange(15)+1 
```

### PLot the relation between the mass and energy

```python
plt.plot(mass, Energy, 'm*', markersize = 8, label='mass energy equivalence')
plt.legend(loc='upper left')
plt.xlabel('Mass in kg')
plt.ylabel('Energy in J')
plt.title('Mass Energy Equivalence')
plt.show()
```

## Activity 2: plot line, quadratic, cubic, and exponential

### Create the set of input values for the equations (interval for plotting)

```python
x = np.linspace(0, 5, 15) # linspace(initial value, final value, even steps)
print(x)
```

If I were to plot the equations with an array as the input values and have them extend from 0 to 20, then the exponential will overpower the linear, quadratic, and cubic. I used linspace to create 20 input values that range from 1 to 5. In this way all four plotted lines are visible and can be compared to each other

### Assigning the x dependence to the following variables. For a linear, quadratic, cubic, and exponential function

```python
linear = x
quad = x**2
cubic = x**3
expo = np.exp(x)
```

### I will plot the functions slowly. One after the other. So that we can see how the next function compares to the previous one in the interval from 0 to 5.

### First, I plot the line  

```python
plt.plot(x, linear, 'k-', markersize = 8, label='linear')
plt.legend(loc='upper left')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### Second, I plot the quadratic with the line for comparison

```python
plt.plot(x, linear, 'k-', markersize = 8, label='linear')
plt.plot(x, quad, 'm-', markersize = 8, label='quadratic')
plt.legend(loc='upper left')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### I introduce the cubic function to the previous line and quadratic function

```python
plt.plot(x, linear, 'k-', markersize = 8, label='linear')
plt.plot(x, quad, 'm-', markersize = 8, label='quadratic')
plt.plot(x, cubic, 'r-', markersize = 8, label='cubic')
plt.legend(loc='upper left')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### For a complete analysis between the functions, I plot all four functions on the same window. This shows how each function relates to their x dependence 

```python
plt.plot(x, linear, 'k-', markersize = 8, label='linear')
plt.plot(x, quad, 'm-', markersize = 8, label='quadratic')
plt.plot(x, cubic, 'r-', markersize = 8, label='cubic')
plt.plot(x, expo, 'c-', markersize = 8, label='exponential')
plt.legend(loc='upper left')
plt.show()
```
